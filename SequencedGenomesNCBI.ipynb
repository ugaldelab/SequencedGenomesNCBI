With the data, now we can generate a few plots.

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
#folder_name = "ncbi_7_27_2014"
```

```python
prok_df = pd.read_csv(folder_name + "/prokaryotes.txt", sep="\t",header=0)
```

```python
prok_df[:1]
```

```python
#Summary by Group (in this case Phylum)
pd.value_counts(prok_df.Group)
```

```python
#First, let's look at the number of genomes per year

#Convert the objects into date
prok_df["Release Date"] = prok_df["Release Date"].apply(pd.to_datetime)
#pd.value_counts(prok_df["Release Date"])

#Now group by year
#pd.value_counts(prok_df["Release Date"])
grouped_year = prok_df.groupby(prok_df['Release Date'].map(lambda x: x.year))
```

```python
grouped_year.count()["TaxID"]
```

```python
#Parameters for plot
from matplotlib import rcParams
rcParams['axes.labelsize'] = 16 #18
rcParams['xtick.labelsize'] = 20 # 15
rcParams['ytick.labelsize'] = 20 #15
rcParams['legend.fontsize'] = 7 # 18
rcParams['font.family'] = 'serif'
rcParams['font.serif'] = ['Computer Modern Roman']
rcParams['text.usetex'] = False
rcParams['figure.figsize'] = 13,10
rcParams['lines.linewidth'] = 3
rcParams['lines.markersize']=15
rcParams['lines.color']="#d95f02"

#Make plot
grouped_year.count()["TaxID"].plot(marker='o', color="#d95f02")
plt.xlim(xmax=2014)
plt.savefig("Genomes_Year_" + folder_name + ".pdf",dpi=500)
```

```python
#Plot classification by phylum

#Parameters for plot
rcParams['axes.labelsize'] = 14 #18
rcParams['xtick.labelsize'] = 24 # 15
rcParams['ytick.labelsize'] = 24 #15
rcParams['legend.fontsize'] = 14 # 18
rcParams['figure.figsize'] = 13,10
rcParams['lines.linewidth'] = 3
rcParams['lines.markersize']=15

pd.value_counts(prok_df.Group)[:10].plot(kind='barh', color="#d95f02",fontsize=15)
plt.tight_layout()
plt.savefig("PhylumCount_" + folder_name + ".pdf",dpi=500)
```

```python
#Group by species
grouped_year = prok_df.groupby(prok_df['Release Date'].map(lambda x: x.year))
grouped_species = prok_df.groupby(prok_df["#Organism/Name"].map(lambda x: " ".join(x.split(" ")[0:2])))


```

```python
#Count of the top species with genomes sequenced
grouped_species.count()["TaxID"].order(ascending=False)[:30].plot(kind='barh', color="#d95f02",fontsize=15)

plt.tight_layout()
plt.savefig("SequencedGenomeSpecies_" + folder_name + ".pdf", dpi=500)

```
